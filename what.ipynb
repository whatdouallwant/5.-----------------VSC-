```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('GooglePlayStore_wild.csv')
```

```python
lan = df.plot(kind='scatter', x='Rating', y='Installs')
lan.set_xlabel('Рейтинг додатків')

lan.set_ylabel('Кількість скачувань')
```

```python
df['Rating'].plot(kind='box')
plt.show()
```

```python
colors = ['#f202fa', '#38fa02', '#02f2fa']
categories = df['Category'].value_counts().nlargest(10).plot(kind='pie', autopct = '%1.f%%', colors = colors)
categories.set_title('Кількість додатків в різних категоріях')
```

```python

categories = df['Category'].value_counts().plot(kind='barh', figsize=(10,6), grid = True)
categories.set_title('Кількість додатків в різних категоріях')
```
